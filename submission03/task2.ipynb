```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
values = np.loadtxt('data/vl5_data.csv')
t_u = np.array(values[:,1])
t_c = np.array(values[:,0])
```

```python
x_mean = np.mean(t_u)
y_mean = np.mean(t_c)
SP_xy = np.sum((t_u - x_mean) * (t_c - y_mean))
SQ_x = np.sum((t_u - x_mean)**2)
alpha_1 = SP_xy / SQ_x
alpha_0 = y_mean - alpha_1 * x_mean

line = alpha_0 + np.multiply(alpha_1,t_u)

plt.plot(t_u, t_c, 'x') 
plt.plot(t_u, line, 'r')
plt.show()
```
